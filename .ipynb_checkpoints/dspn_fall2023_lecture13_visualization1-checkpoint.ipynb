# Wading into the tidyverse...

## Reading in and interpreting a simple dataset

* Dataset contains the body mass index (BMI) of 1786 people, together with the number of steps each of them took on a particular day.
* Stored in two files: one for chronic stroke patients, one for age-matched controls.

## Step #1:
* Reading in the data
    * Coding diagnosis as a variable
    * Merging the data frames

```r
# loading the tidyverse
library(tidyverse)
```

```r
# Setting up our file paths and file names

#  paths
# d_base <- getwd()
d_base <- '/Users/jeremyhogeveen/Dropbox/Fall_2023/teaching/PSY450_650/DSPN_Fall2023_workdir/'

# filename
f_patients <- paste(d_base,'misc/data_visualization_exercise/data_patients.txt',sep='/')
f_controls <- paste(d_base,'misc/data_visualization_exercise/data_controls.txt',sep='/')
```

```r
# Reading in the data and merging into a single DF
# reading in the data
df_patients <- read_tsv(f_patients)
df_patients <- mutate(df_patients,Dx=2)

# more efficient / easier to read way to code in R: Pipes
df_patients <- read_tsv(f_patients) %>% # Can string tidyverse functions together in a chain
    mutate(Dx = 1)
df_controls <- read_tsv(f_controls) %>%
    mutate(Dx = 0)
    
#merging the controls and fecontrols data together
df <- bind_rows(df_patients,df_controls) # tidyverse has a merge function similar to pd, but in this case we're not merging given the two datasets contain independent observations
```

## Step #2:
* Testing whether there is a statistically significant difference in the average number of steps taken by controls and patients
    * Run a t-test
    * Plot in a boxplot

```r
# hypothesis #1:
print(mod_hyp1 <- t.test(steps~Dx,df))

# making sure R knows the Dx var is a factor
# useful for plotting in ggplot
df <- df %>%
    mutate(Dx_factor = as.factor(Dx))

# plotting the data in a scatterplot by group
ggplot(df,aes(x = Dx_factor, y = steps)) + 
    geom_boxplot() +
    theme_minimal()

# therefore...
print('Hypothesis #1 --> Supported by the data')
```

## Step #3:
* Testing that there is a negative correlation between the number of steps and the BMI for stroke patients
    * Run a cor test

```r
# hypothesis #2: 
print(mod_hyp2 <- cor.test(df_patients$steps,df_patients$bmi))
print('Hypothesis #2 --> Supported by the data')
```

## Step #4:
* Testing that there is also a negative correlation between steps and BMI in controls
    * Run a cor test

```r
# hypothesis #3: 
print(mod_hyp3 <- cor.test(df_controls$steps,df_controls$bmi))
print('Hypothesis #3 --> supported by the data')
```

# Step #5:

* Visualizing the data

```r
# plotting the data in a scatterplot by group
plot(plot_bmidata_scatter <- ggplot(df,aes(x = steps, y = bmi, color = Dx_factor)) + geom_point())

plot(plot_bmidata_scatter_wlines <- ggplot(df,aes(x = steps, y = bmi, color = Dx_factor)) + geom_point() + geom_smooth(method=lm))
```
